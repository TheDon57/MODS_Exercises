## Transform data

### Transform variables

Machine learning models cannot work with text categories — they need numbers. To deal with categorical variables in the decision tree we therefore need to dummy encode the variables.

1.  **Define categorical columns:** We need to tell the encoder which columns in the dataset are categorical and need to be converted. Therefore, we define a list of categorical columns.

2. **Create the OneHotEncoder object:**
        
    2.1. We drop the first category from each variable to avoid dummy variable trap. For example: We consider the variable Gender, then:

    * Male → dropped

    * Female → keep column Female (1 if female, else 0)
    
    This prevents redundant columns.
        
    2.2. We need to think about how the OneHotEncoder should handle unknown categories of a Variable as we will perform the dummy encoding on the training set to avoid leakage and then apply it to the test set. If the test set contains a category that did not appear in the training set, the encoder will group it into an “infrequent” bucket, instead of breaking your model.

3. **Fit and transform the training data:** We apply the encoder to the training set `data_training` so that the encoder learns the unique categories for each column and converts the into dummy/one-hot columns.

4. **Create a DataFrame with the encoded variables and join them with the training dataset**: The encoded variables are joined with the original DataFrame. Therefore, we first need to create a separate DataFrame with the econded variables. We can then add this DataFrame to the original training dataset and remove the original columns.

5. **Define target variable:** Now we need to define the variable we want to predict. In this case we want to predict wether a customer will get a credit loan or not - this is represented in the variable `Decision`. Therefore, we drop the column `Decision` from the `data_training` to have a DataFrame of the training set only containing the input variables saving it as `X_train`. We create a second DataFrame called `y_train` that only contains the target variable we want to predict.

6. **Transform the test data:** Lastly, we apply the, on the training data, fitted encoder to the test data and transform the variables of the test data. We then also create a DataFrame with the encoded variables, join them with the test dataset, and define the target variable (just as in the case of the training data).


*Run the code below.*

```python
from sklearn.preprocessing import OneHotEncoder
# Dummy encoding / One-hot encoding

categorical_columns = [
    "Tier",
    "Build_Selfcon",
    "Accommodation_Class",
    "Loan_Type",
    "Gender",
    "Employment_Type",
    "Doc_Proof_Inc",
    "Marital_Status",
    "Employer_Type",
    "Education_Class",
    "Mode_of_origin_class"
]

# Creating the OneHotEncoder object
encoder = OneHotEncoder(drop='first', handle_unknown='infrequent_if_exist', sparse_output=False)

# Fit and transform the training data
encoded_training = encoder.fit_transform(data_training.loc[:,categorical_columns])

# Create a DataFrame with the encoded variables
encoded_training_df = pd.DataFrame(encoded_training, columns=encoder.get_feature_names_out(categorical_columns))

# Join the encoded variables to the original DataFrame and remove the original columns
data_training = data_training.reset_index(drop=True).join(encoded_training_df).drop(columns=categorical_columns)

# define target variable
X_train = data_training.drop(columns="Decision")
y_train = data_training["Decision"]


# Transform the test data
encoded_test = encoder.transform(data_test.loc[:,categorical_columns])
encoded_test_df = pd.DataFrame(encoded_test, columns=encoder.get_feature_names_out(categorical_columns))
data_test = data_test.reset_index(drop=True).join(encoded_test_df).drop(columns=categorical_columns)
X_test = data_test.drop(columns="Decision")
y_test = data_test["Decision"]
```

## Classification Tree

### Grow classification tree on training data

We use the `DecisionTreeClassifier` object to create a classification tree and use the `fit` function to fit it to our training data.
The fit function takes all columns of the `X_train` dataset as the independent variables to predict the outcome. The `y_train` contains the real values of the outcome variable that needs to be predicted.

*Run the code below.*

```python
from sklearn.tree import DecisionTreeClassifier
model_01 = DecisionTreeClassifier(random_state=42).fit(X_train, y_train)
```

### Controlled tree

As stated in the lecture, one potential drawback of decision trees is its tendency to easily __overfit__ if the tree size isn't controlled. Overfitting describes the issue of machine learning algorithms to tailor models to the training data, at the expense of generalization to previously unseen data points.  

One solution to this problem can be to prevent the tree from growing to big by specifying certain parameters in the `DecisionTreeClassifier`.

The most commonly used parameters are:  

* `min_samples_split`: the minimum number of observations that must exist in a node in order for a split to be attempted.  
* `min_samples_leaf`: The minimum number of observations in any terminal node (leaf).  
* `max_depth`: The maximum depth of any node of the final tree, with the root node counted as depth 0.  

*Run the code below.*

```python
model_02 = DecisionTreeClassifier(min_samples_split=60, min_samples_leaf=20, max_depth=5, random_state=42).fit(X_train, y_train)
```

### Pruning

Another option is called __pruning__ and it cuts off insignificant parts of the tree. To achieve this we simply specify the the parameter `ccp_alpha`. It is a complexity parameter, which defines a threshold for any split to be kept in the resulting tree. Roughly speaking it is the minimum improvement of model performance an additional split needs to provide. There are multiple other ways to prune a decision tree.

*Run the code below.*

```python
model_03 = DecisionTreeClassifier(ccp_alpha=0.01, random_state=42).fit(X_train, y_train)
```

### Print trees

*Run the code below.*

## Evaluation

### Model_01

#### Make predictions on test set

Since we are doing classification this time, our evaluation is a little bit different compared to the last weeks.

We use our model of the `DecisionTreeClassifier` trained on the training dataset `X_train` and make predictions on the text dataset `X_test`.

Instead of predicting a numerical outcome variable we are now predicting probabilities which are then transformed into a binary label. In the `where(predictions_test_01[:,1] > 0.5, 1, 0)` statement we can see that we set the threshold at 0.5, so if the predicted probability for a granted credit is greater then 0.5 we are labeling it as granted and otherwise as rejected (0).

You can think about how adjusting the threshold up or down would influence the outcome.


*Run the code below.*

```python
predictions_test_01 = model_01.predict_proba(X_test)
predictions_test_01_binary = np.where(predictions_test_01[:,1] > 0.5, 1, 0)
```

#### Calculate the resulting accuracy and the confusion matrix

Since we switched from regression to classification we are also using different evaluation metrics.

The metric being calculated is the __accuracy__ which denotes how many of our predictions were correct. Therefore, we use the function `accuracy_score` comparing the true values of our output variable `Decision` in the `y_test` and compare it to the **binary predictions** we made `predictions_test_01_binary`.   

In addition to the accuracy we can also print a so called __confusion matrix__ which enables a more detailed interpretation of the errors we made.

There are two possible errors in a binary classification setting:

1. __False Positives (FP):__ The ground truth label is negative (0) but we predict it to be positive (1).
2. __False Negatives (FN):__ The ground truth label is positive (1) but we predict it to be negative (0).

*Run the code below.*

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay

# Calculate the accuracy of the model
accuracy_model_01 = accuracy_score(y_test, predictions_test_01_binary)
print(accuracy_model_01)

# Display the confusion matrix
ConfusionMatrixDisplay.from_predictions(y_test, predictions_test_01_binary)
```

After looking at the confusion matrix we can see that out of the 252 instances that we have in our test set, 197 are positive and 55 are negative. What does that mean for our model evaluation?  
Since four out of five credits were granted, we would achieve an accuracy of 80% by simply predicting positive (1) for every instance. This problem is mainly due to the fact that we have an __unbalanced dataset__, meaning that one class has a lot more instances than the other.   

There are different ways of dealing with this issue. We could for example choose an evaluation metric that takes this imbalance into consideration or we could even adjust our dataset by using methods like __under- or oversampling__.

## ROC and Auc

![ROC_Curve](https://raw.githubusercontent.com/kbrennig/MODS_WS24_25/refs/heads/main/Week_6_exercise/images/ROC_Curve.png)
<br>
Source: Provost & Fawcett (2013)


Another evaluation metric for classification tasks is the __Area under the curve (AUC)__. The AUC is calculated using the __Receiver operating characteristic (ROC)__ curve which is obtained by plotting the true positive rate (TPR) against the false positive rate (FPR) at various threshold settings.

### Plot ROC curve and calculate AUC on test set  

In this case we need the predicted probability of the positive class (1), not the binary prediction. This is required because AUC and ROC curves evaluate model performance across all probability thresholds, not just 0.5.


* `roc_auc_score()` computes how well your model separates class 0 and class 1 using predicted probabilities, not binary predictions. AUC measures how good the model is at ranking positive cases higher than negative cases. This function returns a number between 0.0 and 1.0 - the closer the value to 1, the better the model.

* We can then plot the ROC curve also using the predicted probabilities. The `RocCurveDisplay.from_predictions()` automatically computes the false positive rate (FPR), computes the true positive rate (TPR), generates the ROC curve, and plots it. `Using plot_chance_level=True` we draw a diagonal baseline from from (0, 0) to (1, 1). This line represents random guessing and helps visually compare model quality.
    * If your ROC curve is above this line → your model is better than random
    * If it’s close to the line → the model is weak
    * If it’s below the line → the model predicts the opposite of what it should


*Run the code below.*

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import RocCurveDisplay

# Calculate and Print the AUC score
auc_score = roc_auc_score(y_test, predictions_test_01[:,1])
print("AUC Score:", auc_score)

#plot ROC curve
RocCurveDisplay.from_predictions(y_test, predictions_test_01[:,1], plot_chance_level=True)
```
